```python
from bs4 import BeautifulSoup as soup
import pandas as pd
import requests
import datetime
```

```python
#collecting all links
def links_scrap2(url,n1,n2):
    all_links=pd.DataFrame(columns=['date','link','loc'])
    num=n1 
    while(num<n2):
        num+=1
        home_page=requests.get(url+'?page='+str(num))
        tabel=soup(home_page.content,'html5lib')
        stories=tabel.findAll('div',{'class':'col-xs-12 col-md-4'})
        for i in stories:
            link=i.a['href']
            date=i.p.text.replace('\n','')
            loc=i.find('h5',{'class':'category-name'}).text.replace('\n','')
            all_links=all_links.append({
            'date':date,
            'link':link,
            'loc':loc
            },ignore_index=True)
    
    return all_links
```

```python
#scrapper for photo-slide articles
def photo_article2(url):
    num=1
    base='https://www.amarujala.com'+url+'?pageId='+str(num)
    article_page=requests.get(base)
    article_tabel=soup(article_page.content,'html5lib')
    total_pages=int(article_tabel.find('span',{'class':'total-pic'}).text[5:])
    headline=article_tabel.find('h1')['title']
    content=article_tabel.find('div',{'itemprop':'articleBody'}).text.replace('\n','')
    while(num<total_pages):
        num+=1
        base_url='https://www.amarujala.com'+url+'?pageId='+str(num)
        new_page=requests.get(base_url)
        new_tabel=soup(new_page.content,'html5lib')
        content=content+new_tabel.find('div',{'itemprop':'articleBody'}).text.replace('\n','')
                
    return {'headline':headline,
            'content':content}
```

```python
#scrapper for normal text article
def article_scrap2(url):
    base_url='https://www.amarujala.com'+url
    article_page=requests.get(base_url)
    article_tabel=soup(article_page.content,'html5lib')
    try:
        headline=article_tabel.find('h1').text.replace('\n','')
    except:
        headline=None
    try:
        para=article_tabel.findAll('div',{'class':['article-desc ul_styling','col-xs-12 col-md-4']})
        content=''
        for p in para:
            content=content+p.text.replace('\n','')
    except:
        content=None
    
    article={'headline':headline,
             'content':content}
    return article
    
```

```python
#scrapper for all links collected
def all_articles(links_temp):
    all_news=pd.DataFrame(columns=['headline','content'])
    for i in links_temp.index:
        if(links_temp['link'][i][1:6]=='photo'):
            news=photo_article2(links_temp['link'][i])
        else:
            news=article_scrap2(links_temp['link'][i])
        news['date']=links_temp['date'][i]
        news['loc']=links_temp['loc'][i]
        all_news=all_news.append(news,ignore_index=True)
        print('new article added')
    
    return all_news
    
```
